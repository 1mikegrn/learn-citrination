### Step 1: Read interesting paper, get data

A recent paper by Xue et al. (2017) (http://dx.doi.org/10.1016/j.actamat.2016.12.009) presented results on how optimal experimental design can be used to discover new shape memory alloys.  Shape memory alloys present an attractive alternative to motorized actuators for many applications.  Xue et al. investigated the correlation between the alloy composition and the transition temperature.

In the paper by Xue et al., there is a data table.  I scrape the data table (using Tabula http://tabula.technology/) to create a csv file.

Here's a picture of that data table in Excel.

```python
from IPython.display import Image
Image(filename='./SMA_demo/csv_image.png', width=700)
```

### Step 2: Upload the data to Citrination

Once I have the csv of the data from the table, I upload it at https://citrination.com/csv_to_datasets

Below is a screenshot.  I upload the data, then click the "Next" button in the top right.

```python
from IPython.display import Image
Image(filename='./SMA_demo/csvtodata.png', width=700)
```

For the purpose of this demo, I'm going to treat all the variables as "properties".  While the Titanium, Nickel, Copper, Iron, and Palladium content could also be combined to form a chemical formula, it's more convenient to just treat them as properties for now. 

I name my data set, then click "Save" in the top right.

```python
from IPython.display import Image
Image(filename='./SMA_demo/create_dataset.png', width=700)
```

I now have a searchable data set that I can filter by property.

```python
from IPython.display import Image
Image(filename='./SMA_demo/dataset.png', width=500)
```

### Step 3: Create model

Now I create a DataView using my new data set:   https://citrination.com/data_views

I find my newly created data set under the tab "My Datasets", select it, and click the "Next" button.


```python
from IPython.display import Image
Image(filename='./SMA_demo/dataview_1.png', width=700)
```

- I include all the properties as column headers, then click "Next" again.
- I specify that the Ti, Ni, Cu, Fe, and Pd contents are all real-valued inputs, in the range of 0-100, since these are the percent content of those metals in the alloy.
- I also specify that en (the Pauling electronegativity), ven (the valence electron number), and dor (the Waber-Cromer pseudopotential radii) are also real-valued inputs.  
- I specify that Tp, the transition temperature, is the output.
- Then I click "Next" again.
- Finally, I name and save my Data View:

```python
from IPython.display import Image
Image(filename='./SMA_demo/dataview_2.png', width=700)
```

A data view is now created, and machine learning models start to train.  After waiting about thirty seconds, the training is complete and I can see my new machine learning model!

```python
from IPython.display import Image
Image(filename='./SMA_demo/dataview_3.png', width=700)
```

I can now do things like look at model predictions at new test points (using the Predict tab), visualize my data set using t-SNE (using the Data Summary --> General tabs), or examine my model quality (using the Data Summary --> Estimator tabs).

For example, using the Data Summary --> General tab, I can quickly see that the transition temperature Tp is most strongly correlated with Iron content.  
(Note: all alloys in this data set had 50% Ti content, hence the zero calculated correlation between transition temperature and Ti content.)

```python
from IPython.display import Image
Image(filename='./SMA_demo/correlation.png', width=700)
```
